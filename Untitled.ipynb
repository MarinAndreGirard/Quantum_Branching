```python
import numpy as np

def expectation_value(operator, state):
    # Perform the expectation value calculation
    expectation = np.dot(state.conj(), np.dot(operator, state))

    return expectation.real  # Return the real part of the expectation value

# Example usage:
state = result.states[0].full()
state = np.squeeze(state)
H = H_total.full()

exp_value = expectation_value(H, state)
print("We check if our method works.")
print("Expectation value:", exp_value)
print("Expectation value:", qt.expect(H_total,result.states[0]))
```
